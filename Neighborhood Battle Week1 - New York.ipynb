Define information of interest and filter dataframe

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe_pizza.columns if col.startswith('location.')] + ['id']
dataframe_pizza_filtered = dataframe_pizza.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_pizza_filtered['categories'] = dataframe_pizza_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_pizza_filtered.columns = [column.split('.')[-1] for column in dataframe_pizza_filtered.columns]

dataframe_pizza_filtered
```

analyse each street

```python
# one hot encoding
df_all_onehot = pd.get_dummies(df_all[['categories']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
df_all_onehot['crossStreet'] = df_all['crossStreet'] 

# move neighborhood column to the first column
fixed_columns = [df_all_onehot.columns[-1]] + list(df_all_onehot.columns[:-1])
df_all_onehot = df_all_onehot[fixed_columns]

df_all_onehot.head()
```

Next, let's group rows by categories and by taking the mean of the frequency of occurrence of each category

```python
df_all_grouped = df_all_onehot.groupby('crossStreet').mean().reset_index()
df_all_grouped
```
